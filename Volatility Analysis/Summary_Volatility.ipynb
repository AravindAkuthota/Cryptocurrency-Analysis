```python
# Import dependencies
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Add in column data as lists

coin_name = ["Binance Coin", "Bitcoin", "Cardano", "Dogecoin", "Ethereum",
             "Litecoin", "Monero", "Stellar", "Tether", "XRP"]

year_2018 = [112.4, 68, 110.52, 113.7, 90.13, 88.6, 100.2, 111.18, 17.1, 8.6, 108]

year_2019 = [68.4, 56.3, 73.58, 53.6, 66, 75.9, 66.8, 67, 12.5, 6.3, 58.3]

year_2020 = [81.2, 63.91, 94.44, 82, 83.47, 83.75, 77.4, 94.5, 37.5, 8.88, 98.34]
```

```python
# Add in data as a list of tuples

data = [("Binance Coin", 112.4, 68.4, 81.2),
        ("Bitcoin", 68, 56.3, 63.91),
        ("Cardano", 110.52, 73.58, 94.44),
        ("Dogecoin", 113.7, 53.6, 82),
        ("Ethereum", 90.13, 66, 83.47),
        ("Litecoin", 88.6, 75.9, 83.75),
        ("Monero", 100.2, 66.8, 77.4),
        ("Stellar", 111.18, 67, 94.5),
        ("S&P 500", 17.1, 12.5, 37.5),
        ("Tether", 8.6, 6.3, 8.88),
        ("XRP", 108, 58.3, 98.34)
]
```

```python
# Create summary dataframe of the annual volatility
df = pd.DataFrame(data=data, columns=['coin_name', 'year_2018', 'year_2019', 'year_2020'])
df
```

```python
# Visualize the annual percent volatility as line graphs

year = ["2018", "2019", "2020"]
x = year

a = (112.4, 68.4, 81.2)
b = (68, 56.3, 63.91)
c = (110.52, 73.58, 94.44)
d = (113.7, 53.6, 82)
e = (90.13, 66, 83.47)
f = (88.6, 75.9, 83.75)
g = (100.2, 66.8, 77.4)
h = (111.18, 67, 94.5)
i = (17.1, 12.5, 37.5)
j = (8.6, 6.3, 8.88)
k = (108, 58.3, 98.34)

plt.plot(x, a, label= "Binance Coin" )
plt.plot(x, b, label= "Bitcoin" )
plt.plot(x, c, label= "Cardano" )
plt.plot(x, d, label= "Dogecoin" )
plt.plot(x, e, label= "Ethereum" )
plt.plot(x, f, label= "Litecoin" )
plt.plot(x, g, label= "Monero" )
plt.plot(x, h, label= "Stellar" )
plt.plot(x, i, label= "S&P 500")
plt.plot(x, j, label= "Tether" )
plt.plot(x, k, label= "XRP" )

plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

plt.xlabel("Year", fontsize=16)
plt.ylabel("Annual Percent Volatility", fontsize=16)
plt.title("Annual Percent Volatility (2018-2020) ", fontsize=20)

plt.rcParams["figure.figsize"] = (12,12)

plt.legend(fontsize=14)
plt.savefig("AnnualPercentVolatilityAllCoins.png")

plt.show()
```
