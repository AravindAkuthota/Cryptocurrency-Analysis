```python
# Visualize the annual percent volatility as a line graph

year = ["2018", "2019", "2020"]
x = year

a = (112.4, 68.4, 81.2)
b = (68, 56.3, 63.91)
c = (110.52, 73.58, 94.44)
d = (113.7, 53.6, 82)
e = (90.13, 66, 83.47)
f = (88.6, 75.9, 83.75)
g = (100.2, 66.8, 77.4)
h = (111.18, 67, 94.5)
i = (8.6, 6.3, 8.88)
j = (108, 58.3, 98.4)

plt.plot(x, a, label= "Binance Coin" )
plt.plot(x, b, label= "Bitcoin" )
plt.plot(x, c, label= "Cardano" )
plt.plot(x, d, label= "Dogecoin" )
plt.plot(x, e, label= "Ethereum" )
plt.plot(x, f, label= "Litecoin" )
plt.plot(x, g, label= "Monero" )
plt.plot(x, h, label= "Stellar" )
plt.plot(x, i, label= "Tether" )
plt.plot(x, j, label= "XRP" )

plt.xlabel("Year")
plt.ylabel("Annual Percent Volatility")
plt.title("Annual Percent Volatility (2018-2020) ")

plt.rcParams["figure.figsize"] = (8,10)

plt.legend()
plt.show()
```
