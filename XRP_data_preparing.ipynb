```python
import pandas as pd
import numpy as np
```

```python
path ="./archive/coin_XRP.csv"
df = pd.read_csv(path)
df
```

```python
df = df.drop(columns=['SNo', 'Name', 'Symbol'])
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
start = '2016-07-06'
end = '2021-07-06'
```

```python
fiveyears = (df['Date']>start) & (df['Date']<= end)
```

```python
bitcoin_df = df.loc[fiveyears]
bitcoin_df
```

```python
bitcoin_df.to_csv('XRP(fiveyears).csv')
```
