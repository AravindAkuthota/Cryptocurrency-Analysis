# Moving hourly csv files to MongoDB Atlas Hourly_Cryptocompare DB

```python
# import libraries
import pandas as pd
import numpy as np
import pymongo
from pymongo import MongoClient
import certifi
from mongo2_config import mongo_uri
```

# Storing dataframe in MongoDB (DB name : Hourly_Cryptocompare)

```python
# defining mongodb uri

coins=['binance', 'bitcoin', 'Cardano', 'dodgecoin', 'Ethereum', 'litecoin', 'monero', 'stellar', 'Tether', 'XRP']

client = MongoClient(mongo_uri,tlsCAFile=certifi.where())
db = client['Kaggle_data']
print(client.list_database_names())
```

```python
All_coins_kaggle_df = pd.DataFrame()

for coin in coins:
    dataframe = pd.read_csv(coin+"(fiveyears)"+".csv", index_col=False)
    #append the new df
    All_coins_kaggle_df = All_coins_kaggle_df.append(dataframe, ignore_index = True)
    All_coins_kaggle_df.reset_index(drop=True, inplace=True)

print(All_coins_kaggle_df)
All_coins_kaggle_df
```

```python
# dropping irrelevant columns
All_coins_kaggle_df = All_coins_kaggle_df.drop(columns=['Unnamed: 0'])
All_coins_kaggle_df.head()
```

```python
# exporting the dataframe to csv
All_coins_kaggle_df.to_csv('All_coins_kaggle.csv')
```
